### Check Missing Values
- Count Missing Values

```python
df.isnull().sum().sort_values(ascending=False)
```

- Percentage of Missing Values (IMPORTANT)

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent.sort_values(ascending=False)
```

```python
# Fill energy-kj using kcal conversion
df['energy-kj_value'] = df['energy-kj_value'].fillna(
    df['energy-kcal_value'] * 4.184
)

# Fill numeric columns with median
numeric_cols = [
    'fiber_value',
    'nutrition-score-fr',
    'salt_value',
    'sodium_value',
    'saturated-fat_value',
    'sugars_value',
    'carbohydrates_value',
    'energy-kcal_value',
    'proteins_value',
    'fat_value',
    'fruits-vegetables-nuts-estimate-from-ingredients_100g'
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill nova group with mode
df['nova-group'] = df['nova-group'].fillna(df['nova-group'].mode()[0])

# Fill brand
df['brand'] = df['brand'].fillna("Unknown")

# Drop rows where product_name missing
df = df.dropna(subset=['product_name'])
```

## Feature Engineering
- sugar_to_carb_ratio

```python
df['sugar_to_carb_ratio'] = df['sugars_value'] / df['carbohydrates_value']

df['sugar_to_carb_ratio'] = df.apply(
    lambda x: x['sugars_value'] / x['carbohydrates_value']
    if x['carbohydrates_value'] != 0 else 0,
    axis=1
)
```

```python
df
```

### calorie_category

Chocolate typically:

< 300 kcal → Low

300–500 → Moderate

500 → High

```python
def calorie_category(x):
    if x < 300:
        return 'Low'
    elif x < 500:
        return 'Moderate'
    else:
        return 'High'

df['calorie_category'] = df['energy-kcal_value'].apply(calorie_category)
```

- sugar_category

```python
def sugar_category(x):
    if x < 10:
        return 'Low Sugar'
    elif x < 25:
        return 'Moderate Sugar'
    else:
        return 'High Sugar'

df['sugar_category'] = df['sugars_value'].apply(sugar_category)
```

#### is_ultra_processed

In OpenFoodFacts:

NOVA = 4 → Ultra-processed

```python
df['is_ultra_processed'] = df['nova-group'].apply(
    lambda x: 'Yes' if x == 4 else 'No'
)
```
